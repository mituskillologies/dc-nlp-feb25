##### Text - Preprocessing

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
swords = stopwords.words('english')
```

```python
sent = 'Hello friends! How are you? I like Python programming.'
```

```python
# tokenize
tokens1 = word_tokenize(sent)
```

```python
tokens1
```

```python
# remove the punctuation
tokens2 = [token for token in tokens1 if token.isalpha()]
```

```python
tokens2
```

```python
# remove the stopwords
tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
```

```python
tokens3
```

```python
# apply the stemming
ps = PorterStemmer()
tokens4 = [ps.stem(token) for token in tokens3]
```

```python
tokens4
```

```python
def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4
```

```python
clean_text(sent)
```

```python
clean_text(sent1)
```

##### Vectorization

```python
sentence_1 = 'It is a good practice for us.'
sentence_2 = 'It was also good to know about it.'
```

```python
# bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvt = CountVectorizer()
```

```python
bag = cvt.fit_transform([sentence_1, sentence_2])
```

```python
pd.DataFrame(bag.toarray(), columns= cvt.get_feature_names_out())
```

```python
# TF*IDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tvect = TfidfVectorizer()
```

```python
vect = tvect.fit_transform([sentence_1, sentence_2])
```

```python
pd.DataFrame(vect.toarray(), columns= tvect.get_feature_names_out())
```

```python
tvect = TfidfVectorizer(analyzer=clean_text)
```

```python
x_vect = tvect.fit_transform(x)
```

```python
x_vect.shape
```

```python
x_new = pd.DataFrame(x_vect.toarray(), columns= tvect.get_feature_names_out())
```

```python
x_new
```

```python
x_new.shape
```

##### Cross - Validation

```python
# import the function
from sklearn.model_selection import train_test_split
```

```python
# split the data
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=0)
```

```python
x_train.shape, x_test.shape
```

##### Build the model

```python
# import the class of any ML algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
# create the object
ml = RandomForestClassifier(random_state=0)
```

```python
# train the algorithm
ml.fit(x_train, y_train)
```

##### Predict the result on unseen (test) data

```python
y_pred = ml.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

##### Evaluate the algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred) * 100
```

```python
sum(y_test == y_pred) / len(y_pred) * 100
```
