```python
# import the class
from bs4 import BeautifulSoup
```

```python
# create the soup object
soup = BeautifulSoup(text, 'html.parser')
```

```python
data = soup.get_text()
```

```python
print(data)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens1 = word_tokenize(data)
```

```python
tokens1
```

```python
from nltk import FreqDist
```

```python
freq = FreqDist(tokens1)
```

```python
freq.plot(20);
```

##### Remove the punctuations

```python
tokens2 = [token for token in tokens1 if token.isalpha()]
```

```python
tokens2
```

##### Stopwords Removal

```python
from nltk.corpus import stopwords
```

```python
swords = stopwords.words('english')
```

```python
swords
```

```python
tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
```

```python
tokens3
```

```python
freq = FreqDist(tokens3)
freq.plot(20)
```

```python
freq.most_common(5)
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
tokens4 = [ps.stem(token) for token in tokens3]
```

```python
freq = FreqDist(tokens4)
freq.plot(20)
```

```python
freq.most_common(5)
```

```python
url = urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
text = url.read()
text = text.decode('utf-8')
soup = BeautifulSoup(text, 'html.parser')
data = soup.get_text()
tokens1 = word_tokenize(data)
tokens2 = [token for token in tokens1 if token.isalpha()]
tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
tokens4 = [ps.stem(token) for token in tokens3]
freq = FreqDist(tokens4)
freq.most_common(5)
```
